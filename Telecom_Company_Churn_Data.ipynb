### 3.3 Creating Derived Metrics

```python
# Assessing column age on network

churn_v6['aon'].describe()
```

```python
# # Defining a function to create buckets of the age on network column

def aon_buckets(x) :
    if x['aon'] <= 450 :
        return '0-450'
    elif (x['aon'] > 450) & (x['aon'] <= 900):
        return '451-900'
    elif (x['aon'] > 900) & (x['aon'] <= 2000) :
        return '901-2000'
    elif x['aon'] > 2000 :
        return '2001-4321'
churn_v6['aon_buckets'] = churn_v6.apply(lambda x:aon_buckets(x),axis = 1)
churn_v6.shape
```

```python
# Dropping the aon columns since we have derived a column from it

churn_v6 = churn_v6.drop(columns = ['aon'])
churn_v6.shape
```

```python
# Creating a column arpu_diff as difference between action phase(8) and mean of good phase(6,7)

churn_v6['arpu_diff'] = churn_v6['arpu_8'] - (churn_v6['arpu_6']+churn_v6['arpu_7'])/2
```

```python
# Creating a column to flag the type of customer e.g only calling, only Internet, both Calling and Internet and None

churn_v6['tot_calls_mou'] = churn_v6['total_ic_mou_6'] + churn_v6['total_og_mou_6'] + churn_v6['total_ic_mou_7'] + churn_v6['total_og_mou_7'] + churn_v6['total_ic_mou_8'] + churn_v6['total_og_mou_8']
churn_v6['tot_data_mb'] =  churn_v6['vol_2g_mb_6'] + churn_v6['vol_3g_mb_6'] + churn_v6['vol_2g_mb_7'] + churn_v6['vol_3g_mb_7'] + churn_v6['vol_2g_mb_8'] + churn_v6['vol_3g_mb_8']

def uType(x):
    if (x['tot_calls_mou'] == 0 and x['tot_data_mb'] > 0):
        userType = "Internet Service Only"
    elif (x['tot_calls_mou'] > 0 and x['tot_data_mb'] == 0):
         userType = "Call Service Only"
    elif (x['tot_calls_mou'] > 0 and x['tot_data_mb'] > 0):
        userType = "Both"
    else:
        userType = "None"
    return userType

churn_v6['UserType']= churn_v6.apply(uType,axis=1)
```
